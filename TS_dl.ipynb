```python
def clean_data(file="./train.csv",test=0):
    df=pd.read_csv(file)
    l=0
    d={}
    for name in range(len(df["Name"])):
        for i in range(len(df["Name"][name])):
            if df["Name"][name][i]==',':
                if df["Name"][name][:i] not in d.keys():
                    d[df["Name"][name][:i]]=l
                    l+=1
                    df["Name"][name]=l
                else:
                    df["Name"][name]=d[df["Name"][name][:i]]
                break
    df["Sex"]=(df["Sex"]=="male")*1
    d={"S":1,"C":2,"Q":3,np.nan:0}
    for i in range(len(df)):
        df["Embarked"][i]=d[df["Embarked"][i]]
    df=df.drop(columns=["Ticket","Cabin"])
    df=df.fillna(0)
    df['per_person']=pd.Series(df["Fare"]/(1+df["SibSp"]+df["Parch"]),index=df.index)
    if(test==0):
        labels=df["Survived"].values
        df=df.drop(columns=["PassengerId","Survived"])
        for i in df.columns:
            sd=np.max(df[i])-np.min(df[i])
            p=np.mean(df[i])
            df[i]=(df[i]-p)/sd
        return df.values,labels
    else:
        for i in df.columns:
            sd=np.max(df[i])-np.min(df[i])
            p=np.mean(df[i])
            df[i]=(df[i]-p)/sd
        return df.values
```

```python
x1=x[:700]
x2=x[700:]
labels1=labels[:700]
labels2=labels[700:]
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def h(theta,x,b):
    y=np.dot(x,theta)+b
    return sigmoid(y)
def sigmoid_deff(x):
    return sigmoid(x)*(1-sigmoid(x))
def compute_cost(y,y1,m,regu):
    return (-np.sum(xlogy(y,y1)+xlogy(1-y,1-y1))/m)+regu*np.sum()
def model(X,y,num_hidden=3,sizes=[9,9,9,9,1],num_iters=500,lr=0.01,regu=1):
    num_inputs=X.shape[1]
    m=X.shape[0]
    X=X.T
    params={"Z0":X,"A0":X}
    for i in range(num_hidden+1):
        params["W"+str(i+1)]=np.random.randn(sizes[i+1],sizes[i])*0.01
        params["b"+str(i+1)]=np.zeros([sizes[i+1],1])
    for j in range(num_iters):
        for i in range(num_hidden+1):
            params["Z"+str(i+1)]=np.dot(params["W"+str(i+1)],params["Z"+str(i)])+params["b"+str(i+1)]
            params["A"+str(i+1)]=sigmoid(params["Z"+str(i+1)])
        params["dZ"+str(num_hidden+1)]=params["A"+str(i+1)]-y
        params["dW"+str(num_hidden+1)]=np.dot(params["dZ"+str(num_hidden+1)],params["A"+str(num_hidden)].T)/m
        params["db"+str(num_hidden+1)]=np.sum(params["dZ"+str(num_hidden+1)],axis=1,keepdims=True)/m
        for i in range(num_hidden,0,-1):
            params["W"+str(i+1)]=params["W"+str(i+1)]-lr*params["dW"+str(i+1)]
            params["b"+str(i+1)]=params["b"+str(i+1)]-lr*params["db"+str(i+1)]
            params["dZ"+str(i)]=np.dot(params["W"+str(i+1)].T,params["dZ"+str(i+1)])*sigmoid_deff(params["Z"+str(i)])
            params["dW"+str(i)]=np.dot(params["dZ"+str(i)],params["A"+str(i-1)].T)/m
            params["db"+str(i)]=np.sum(params["dZ"+str(i)],axis=1,keepdims=True)/m
        params["W"+str(1)]=params["W"+str(1)]-lr*params["dW"+str(1)]
        params["b"+str(1)]=params["b"+str(1)]-lr*params["db"+str(1)]
        if j%100==0:
            print("The cost after iter ",str(j),": ",str(compute_cost(y,params["A"+str(num_hidden+1)],m)))
            #print(params["A4"])
    l=0
    for i in range(m):
        if params["A4"][i][0]>0.5:
            params["A4"][i][0]=1
        else:
            params["A4"][i][0]=0
        if params["A4"][i][0]==y[i][0]:
            l+=1
    print(l/m)
    return params
    
```

```python
params=model(x1,labels1)
```

```python
def predict(params,X,y,num_hidden=3):
    num_inputs=X.shape[1]
    m=X.shape[0]
    X=X.T
    params["Z0"]=X
    for i in range(num_hidden+1):
        params["Z"+str(i+1)]=np.dot(params["W"+str(i+1)],params["Z"+str(i)])+params["b"+str(i+1)]
        params["A"+str(i+1)]=sigmoid(params["Z"+str(i+1)])
    l=0
    for i in range(m):
        if params["A4"][i][0]>0.5:
            params["A4"][i][0]=1
        else:
            params["A4"][i][0]=0
        if params["A4"][i][0]==y[i][0]:
            l+=1
    print(l/m)
```
