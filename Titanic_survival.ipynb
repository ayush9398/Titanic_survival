```python
x=traindf.values
b=np.random.rand(1,1)
theta=np.random.rand(x.shape[1],1)
m=x.shape[0]
def sigmoid(x):
    return 1/(1+np.exp(-x))
def h(theta,x,b):
    y=np.dot(x,theta)+b
    return sigmoid(y)

def J(theta,x,b,y):
    return np.sum(-y*np.log(h(theta,x,b)-(1-y)*np.log(1-h(theta,x,b))))/m

def grad_desc(lr,theta,x,y,b):
    theta=theta-lr*np.dot(x.T,(h(theta,x,b)-y))/m
    b=b-lr*np.sum(h(theta,x,b)-y)/m
    return theta,b

def train(x,y,theta,b,iterations=20000,lr=0.001):
    for i in range(iterations):
        theta,b=grad_desc(lr,theta,x,y,b)
        if i%500==0:
            print("Cost: ",J(theta,x,b,y))
    return (theta,b)
def predict(model,x):
    theta=model[0]
    b=model[1]
    x=pd.read_csv(x)
    p=x["PassengerId"]
    for name in range(len(x['Name'])):
        for i in range(len(x['Name'][name])):
            if x['Name'][name][i]==',':
                x['Name'][name]=x['Name'][name][:i]
                break
    for i in range(len(x)):
        if x["Sex"][i]=="male":
            x["Sex"][i]=1
        else:
            x["Sex"][i]=0
    for i in range(len(x['Embarked'])):
        if x['Embarked'][i]=="S":
            x['Embarked'][i]=0
        elif x['Embarked'][i]=="Q":
            x['Embarked'][i]=1
        elif x['Embarked'][i]=="C":
            x['Embarked'][i]=2

    n=x['Name'].unique()
    d={}
    for i in range(len(n)):
        d[n[i]]=i
    for i in range(len(x)):
        x['Name'][i]=d[x['Name'][i]]
    x=x.drop(['Ticket','Cabin'],axis=1)
    x['per_person']=pd.Series(x["Fare"]/(1+x["SibSp"]+x["Parch"]),index=x.index)
    for i in x.columns[1:]:
        sd=np.max(x[i])-np.min(x[i])
        m=np.mean(x[i])
        x[i]=(x[i]-m)/sd
    x=x.fillna(0)
    x=x.drop(['PassengerId'],axis=1)
    x=x.values
    y1=h(theta,x,b)
    n=0
    d=0
    csv_data=[]
    for i in range(len(y1)):
        if(y1[i]>0.5):
            y1[i]=1
        else:
            y1[i]=0
        csv_data.append(np.array([int(p[i]),int(y1[i])]))
    return csv_data

model=train(x,labels,theta,b,80000,0.01)
```
